```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("semi.csv")

q1 = data[["Producer", "Employed"]]
q2 = data[["Producer", "Relative"]]
q3 = data[["Relative", "Industrial"]]
```

```python
q1
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(q1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeansOne = KMeans(n_clusters=4, precompute_distances="auto", n_jobs=-1)
clusterOne = kmeansOne.fit_predict(q1)
print(clusterOne)
plt.scatter(q1["Producer"], q1["Employed"],c=kmeansOne.labels_.astype(float))
plt.scatter(kmeansOne.cluster_centers_[:, 0], kmeansOne.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel("Production Index")
plt.ylabel("Employment Index")
plt.show()
```
